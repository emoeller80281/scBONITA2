```python
import os
import pandas as pd

dataset_dir = "/home/emoeller/github/scBONITA2/input/atherosclerosis_dataset"

participant_metadata_dfs = []
participant_processed_gene_expression_dfs = []
for file in os.listdir(dataset_dir):
    if file.endswith(".csv"):
        if "metadata" in file:
            metadata_df = pd.read_csv(os.path.join(dataset_dir, file), index_col=0)
            metadata_df = metadata_df[["Atherosclerosis Status"]]
            participant_metadata_dfs.append(metadata_df)
        elif "processed_gene_expression" in file:
            processed_gene_expression_df = pd.read_csv(os.path.join(dataset_dir, file), index_col=0)
            participant_processed_gene_expression_dfs.append(processed_gene_expression_df)
            
all_metadata_df = pd.concat(participant_metadata_dfs, axis=0)
all_metadata_df = all_metadata_df[~all_metadata_df.index.duplicated(keep='first')]

all_processed_gene_expression_df = pd.concat(participant_processed_gene_expression_dfs, axis=1)
all_processed_gene_expression_df.fillna(0, inplace=True)

all_processed_gene_expression_df.to_csv("/home/emoeller/github/scBONITA2/input/atherosclerosis_dataset/combined_atherosclerosis_scRNAseq.csv")
all_metadata_df.to_csv("/home/emoeller/github/scBONITA2/input/atherosclerosis_dataset/combined_atherosclerosis_metadata.csv")
```
